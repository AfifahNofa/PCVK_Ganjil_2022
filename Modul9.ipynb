7. Buatlah program implementasi operasi opening dengan Structuring Element
berbentuk square 3 x 3 tanpa dan dengan menggunakan library morphology dari
openCV, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar
fingerprint.png)

```python
#No7
#Buatlah program implementasi operasi opening dengan Structuring Element berbentuk square 3 x 3 tanpa menggunakan library morphology dari openCV
# Membaca gambar
img7 = cv2.imread('/content/drive/MyDrive/PCVK/Images/fingerprint.png', 0)

# Acquire size of the image
m, n = img7.shape

# Show the image
plt.figure(figsize=(12, 4))

# Define the structuring element (SE) - Square 3x3
k = 3
SE = np.ones((k, k), dtype=np.uint8)
constant = (k - 1) // 2

# Define new image for erosion
imgErode = np.zeros((m, n), dtype=np.uint8)

# Erosion without using inbuilt cv2 function for morphology
for i in range(constant, m - constant):
    for j in range(constant, n - constant):
        temp = img7[i - constant:i + constant + 1, j - constant:j + constant + 1]
        product = temp * SE
        imgErode[i, j] = np.min(product)

# Show the original image
plt.subplot(131)
plt.imshow(img7, cmap="gray")
plt.title('Citra Awal')

# Show the eroded image
plt.subplot(132)
plt.imshow(imgErode, cmap="gray")
plt.title('Citra Hasil Erosi')

# Define new image for opening
imgOpen = np.zeros((m, n), dtype=np.uint8)

# Dilation operation without using inbuilt cv2 function
for i in range(constant, m - constant):
    for j in range(constant, n - constant):
        temp = imgErode[i - constant:i + constant + 1, j - constant:j + constant + 1]
        product = temp * SE
        imgOpen[i, j] = np.max(product)

# Show the opened image
plt.subplot(133)
plt.imshow(imgOpen, cmap="gray")
plt.title('Citra Hasil Opening')

plt.show()
```

```python
#No.7
#Buatlah program implementasi operasi opening dengan Structuring Element berbentuk square 3 x 3 dengan menggunakan library morphology

# Membaca gambar
img7 = cv2.imread('/content/drive/MyDrive/PCVK/Images/fingerprint.png', 0)

# Acquire size of the image
m, n = img7.shape

# Show the image
plt.figure(figsize=(12, 4))

# Define the structuring element (SE) - Square 3x3
k = 3
SE = np.ones((k, k), dtype=np.uint8)

# Erosion using OpenCV's morphology function
imgErode = cv2.erode(img7, SE, iterations=1)

# Show the original image
plt.subplot(131)
plt.imshow(img7, cmap="gray")
plt.title('Citra Awal')

# Show the eroded image
plt.subplot(132)
plt.imshow(imgErode, cmap="gray")
plt.title('Citra Hasil Erosi')

# Define new image for opening
imgOpen = cv2.dilate(imgErode, SE, iterations=1)

# Show the opened image
plt.subplot(133)
plt.imshow(imgOpen, cmap="gray")
plt.title('Citra Hasil Opening')

plt.show()
```

8. Buatlah program implementasi operasi closing dengan Structuring Element berbentuk
square 3 x 3 tanpa dan dengan menggunakan library morphology dari openCV,
sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar
fingerprint.png)

```python
#No.8
#Buatlah program implementasi operasi closing dengan Structuring Element berbentuk square 3 x 3 tanpa menggunakan library morphology dari openCV

# Membaca gambar
img7 = cv2.imread('/content/drive/MyDrive/PCVK/Images/fingerprint.png', 0)

# Acquire size of the image
m, n = img7.shape

# Show the image
plt.figure(figsize=(12, 4))

# Define the structuring element (SE) - Square 7x7
k = 3
SE = np.ones((k, k), dtype=np.uint8)
constant = (k - 1) // 2

# Define new image for dilation
imgDilate = np.zeros((m, n), dtype=np.uint8)

# Dilation without using inbuilt cv2 function for morphology
for i in range(constant, m - constant):
    for j in range(constant, n - constant):
        temp = img7[i - constant:i + constant + 1, j - constant:j + constant + 1]
        product = temp * SE
        imgDilate[i, j] = np.max(product)

# Show the original image
plt.subplot(131)
plt.imshow(img7, cmap="gray")
plt.title('Citra Awal')

# Show the dilated image
plt.subplot(132)
plt.imshow(imgDilate, cmap="gray")
plt.title('Citra Hasil Dilasi')

# Define new image for closing (dilasi diikuti erosi)
imgClose = np.zeros((m, n), dtype=np.uint8)

# Erosion without using inbuilt cv2 function for morphology
for i in range(constant, m - constant):
    for j in range(constant, n - constant):
        temp = imgDilate[i - constant:i + constant + 1, j - constant:j + constant + 1]
        product = temp * SE
        imgClose[i, j] = np.min(product)

# Show the closed image (dilasi diikuti erosi)
plt.subplot(133)
plt.imshow(imgClose, cmap="gray")
plt.title('Citra Hasil Closing (Dilasi diikuti Erosi)')

plt.show()
```

```python
#No.8
#Buatlah program implementasi operasi closing dengan Structuring Element berbentuk square 3 x 3 dengan menggunakan library morphology dari openCV

# Membaca image
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/fingerprint.png', 0)

# Membaca ukuran gambar
m, n = img.shape

# Menampilkan gambar
plt.figure(figsize=(12, 4))

# Mendefinisikan Structuring Element (SE) berbentuk Square 3x3
k = 3
SE = np.ones((k, k), dtype=np.uint8)

# Melakukan operasi dilasi menggunakan library morphology dari OpenCV
imgDilate = cv2.dilate(img, SE)

# Menampilkan gambar asli
plt.subplot(131)
plt.imshow(img, cmap="gray")
plt.title('Citra Awal')

# Menampilkan gambar hasil dilasi
plt.subplot(132)
plt.imshow(imgDilate, cmap="gray")
plt.title('Citra Hasil Dilasi')

# Melakukan operasi closing menggunakan library morphology dari OpenCV
imgClose = cv2.morphologyEx(img, cv2.MORPH_CLOSE, SE)

# Menampilkan gambar hasil closing
plt.subplot(133)
plt.imshow(imgClose, cmap="gray")
plt.title('Citra Hasil Closing (Dilasi diikuti Erosi)')

plt.show()
```

9. Buatlah program implementasi operasi dilasi dengan Structuring Element berbentuk
Cross 3 x 3 dan 5 x 5, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan
gambar j.png)

```python

# Membaca gambar
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/j.png')

# Mendefinisikan kernel SE berbentuk Cross 3x3
kernel3x3 = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))

# Dilasi dengan SE Cross 3x3
dilation3x3 = cv2.dilate(img, kernel3x3, iterations=1)

# Mendefinisikan kernel SE berbentuk Cross 5x5
kernel5x5 = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))

# Dilasi dengan SE Cross 5x5
dilation5x5 = cv2.dilate(img, kernel5x5, iterations=1)

# Menampilkan gambar awal dan hasil dilasi Cross 3x3
plt.figure(figsize=(12, 10))
print("SED = Cross 3")
plt.subplot(231), plt.imshow(img, cmap='gray'), plt.title('Citra Awal')
plt.subplot(232), plt.imshow(dilation3x3, cmap='gray'), plt.title('Citra Hasil Dilasi (Cross 3x3)')

# Menampilkan gambar awal dan hasil dilasi Cross 5x5
print("SED = Cross 5")
plt.subplot(234), plt.imshow(img, cmap='gray'), plt.title('Citra Awal')
plt.subplot(235), plt.imshow(dilation5x5, cmap='gray'), plt.title('Citra Hasil Dilasi (Cross 5x5)')

plt.show()
```

11. Buatlah program implementasi operasi erosi dengan Structuring Element berbentuk

(Gunakan gambar j.png)
